```python
#Write a program which creates nine new folders – one for sequences between 100 and 199 bases long, one for sequences between 200 and 299 bases long
#first, from terminal cd into the folder with the files you need (cd /Users/kel/Desktop/bio_programming/exercises and examples/working_with_the_filesystem/exercises)

#1) Read all the DNA sequences

import os

#create a new folder for each bin

for bin_lower in range(100,1000,100):
    bin_upper = bin_lower + 99
    bin_folder_name = str(bin_lower) + "_" + str(bin_upper)
    os.mkdir(bin_folder_name)

# create a variable to hold the sequence number
seq_number = 1

#Get a list of all the files in the folder using os.listdir(.) the . lists ALL files
#To only read DNA sequences from files that have the right file name extension use .endswith 
#print to check work

for file_name in os.listdir("."):
    if file_name.endswith(".dna"):
        print("reading sequences from " + file_name)
        #open the files that end with .dna
        dna_file = open(file_name)
        
        #now look at each line at a time to calculate the length of the DNA
        for line in dna_file:
            #remove line breaks
            dna = line.rstrip("\n")
            #calculate the length of the dna
            dna_length=len(dna)
            #print to check work
            print("found a dna sequence with length " + str(dna_length))
            
            # go through each bin and check if the sequence belongs in it
            for bin_lower in range(100,1000,100):
                bin_upper = bin_lower + 99
                if dna_length >= bin_lower and dna_length < bin_upper:
                    print("bin is " + str(bin_lower) + " to " + str(bin_upper))
                    
                    #Figure out in which bin seq belongs to. create the new folders, and write each DNA sequence to the appropriate one.
                    # once we know the correct bin, write out the sequence
                    
                    bin_folder_name = str(bin_lower) + "_" + str(bin_upper)
                    output_path = bin_folder_name + '/' + str(seq_number) + '.dna'
                    output = open(output_path, "w")
                    output.write(dna)
                    output.close()
                    
                    # increment the sequence number
                    seq_number = seq_number+1
```
